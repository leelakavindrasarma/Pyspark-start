```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum, avg, max, count

# Create a SparkSession
spark = SparkSession.builder.appName("GroupByExample").getOrCreate()

# Create a sample DataFrame
data = [("Sales", "NY", 5000),
        ("Sales", "NY", 6000),
        ("Sales", "CA", 8000),
        ("HR", "NY", 3000),
        ("HR", "CA", 4000),
        ("IT", "CA", 6000),
        ("IT", "NY", 7000)]
df=spark.createDataFrame(data,("dep","ST","salary"))
df.show()
group = df.groupby("dep").agg(sum("salary").alias("full_salary"))
group.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.appName("RightOverExample").getOrCreate()

# Create sample DataFrames
df2 = spark.createDataFrame([(21, "A"), (24, "B"), (83, "C")], ["id", "value"])
df1 = spark.createDataFrame([(23, "X"), (32, "Y"), (49, "Z")], ["id", "other_value"])
df11=df2.orderBy(col("id").desc())
df22=df1.orderBy("id")
df11.show()
df22.show()
```
